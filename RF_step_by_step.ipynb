## Random foorests

### 1 Bootstrap sampling the training data

```
N = 100 #number of tree in the forest
```

```
training_data = list() 
for i in range(N): #ต้นไม้เเต่ละต้นเห็น data = 80%
    X_sp, X_dp, y_sp, y_dp = train_test_split(X_train, y_train, test_size=0.2, random_state = i+6)
    training_data.append({'x':X_sp,'y':y_sp})
```

## 2 Train separately

```
classifier_forest = list() # Train ข้อมูลโดยการวนลูปแล้ว เก็บในตัวแปร classifier_forest
for i in range(N):
    tree = DecisionTreeClassifier(max_features='sqrt' ,random_state= i+2)
    tree.fit(training_data[i]['x'], training_data[i]['y'])
    classifier_forest.append(tree)
```

### 3 To predict, aggregating

```
prediction = list() # เก็บค่าที่ทำนายได้ในทุกโมเดล ในตัวแปร prediction
for i in range(N):
    pred = classifier_forest[i].predict(X_test)
    #print(pred)
    prediction.append(pred)
```

```
prediction # เรียกดูข้อมูลของตัวแปร prediction
```

```
from scipy import stats #import
import numpy as np
```

```
m = stats.mode(np.array(prediction)) # สร้างตัวแปร m เพื่อเก็บค่าฐานนิยมของตัวแปร prediction
```

```
m #เรียกดูข้อมูล
```

## 4 Evaluation

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(y_test, m.mode.squeeze()) # คำนวณเเละเเสดงค่าความแม่นยำ
```
